Task 2: Apply dimension reduction techniques (10 marks)

High-dimensional data can lead to slow training times, overfitting, and increased memory usage. The train dataset contains 5000 TF-IDF features. In this task, you are to compare two distinct approaches for handling high-dimensional text data: dimensionality reduction via PCA (Principal Component Analysis) versus direct feature selection by simply restricting the number of top TF-IDF features used.


```python
import pandas as pd
import sklearn

df_train = pd.read_csv('train.csv',sep='\t') # fix the rows that have extra commas issue
df_test = pd.read_csv('test.csv',sep='\t')   
print(df_train.head())
print(df_test.head())
print(df_train.shape)
print(df_test.shape)
```

```python
# convert abstract column to 5000 TF-IDF features

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)

X_tfidf = vectorizer.fit_transform(df_train['abstract'])

print(X_tfidf.shape)
# print(X_tfidf[0])
```

```python
#  labels for trainset
y = df_train['label_id']
print(y.shape)
```

```python
#  for test set, use the same vectorizer fitted on training data

vectorizer.fit(df_train['abstract']) 
X_test_tfidf = vectorizer.transform(df_test['abstract'])
print(X_test_tfidf.shape)
```

2a. Code implementation to reduce the feature space to 2000, 1000, 500, and 100 features using two separate methods:

Feature Selection: Re-computing the TF-IDF matrix while restricting the vocabulary to the top N features.

```python
# next portion will follow conversion of abstract column for both train and test set to N number of features, both fitted on training data

# convert abstract column to 2000 TF-IDF features

vectorizer_2k = TfidfVectorizer(max_features = 2000)
X_tfidf_2k = vectorizer_2k.fit_transform(df_train['abstract'])
print(X_tfidf_2k.shape)
X_test_tfidf_2k = vectorizer_2k.transform(df_test['abstract'])
print(X_test_tfidf_2k.shape)


# convert abstract column to 1000 TF-IDF features

vectorizer_1k = TfidfVectorizer(max_features = 1000)
X_tfidf_1k = vectorizer_1k.fit_transform(df_train['abstract'])
print(X_tfidf_1k.shape)
X_test_tfidf_1k = vectorizer_1k.transform(df_test['abstract'])
print(X_test_tfidf_1k.shape)


# convert abstract column to 500 TF-IDF features

vectorizer_500 = TfidfVectorizer(max_features = 500)
X_tfidf_500 = vectorizer_500.fit_transform(df_train['abstract'])
print(X_tfidf_500.shape)
X_test_tfidf_500 = vectorizer_500.transform(df_test['abstract'])
print(X_test_tfidf_500.shape)


# convert abstract column to 100 TF-IDF features

vectorizer_100 = TfidfVectorizer(max_features = 100)
X_tfidf_100 = vectorizer_100.fit_transform(df_train['abstract'])
print(X_tfidf_100.shape)
X_test_tfidf_100 = vectorizer_100.transform(df_test['abstract'])
print(X_test_tfidf_100.shape)
```

Dimension Reduction: Applying PCA (or its sparse equivalent) to project the original 5000 TF-IDF features down to N components. (Note: You are allowed to use the sklearn packages for both TF-IDF and PCA/TruncatedSVD for this task.)

```python
# this portion will follow with applying PCA on the 5000 T-IDF features, we will be using only the train set for this one to get the N features which will be applied on the test set

from sklearn.decomposition import TruncatedSVD

svd_2k = TruncatedSVD(n_components=2000)
X_pca_2k = svd_2k.fit_transform(X_tfidf)
X_test_pca_2k = svd_2k.transform(X_test_tfidf)
print(X_pca_2k.shape)
print(X_test_pca_2k.shape)

svd_1k = TruncatedSVD(n_components=1000)
X_pca_1k = svd_1k.fit_transform(X_tfidf)
X_test_pca_1k = svd_1k.transform(X_test_tfidf)
print(X_pca_1k.shape)
print(X_test_pca_1k.shape)

svd_500 = TruncatedSVD(n_components=500)
X_pca_500 = svd_500.fit_transform(X_tfidf)
X_test_pca_500 = svd_500.transform(X_test_tfidf)
print(X_pca_500.shape)
print(X_test_pca_500.shape)

svd_100 = TruncatedSVD(n_components=100)
X_pca_100 = svd_100.fit_transform(X_tfidf)
X_test_pca_100 = svd_100.transform(X_test_tfidf)
print(X_pca_100.shape)
print(X_test_pca_100.shape)
```
